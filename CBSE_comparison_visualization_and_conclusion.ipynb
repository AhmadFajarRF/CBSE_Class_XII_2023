# comparison visualization

```python
import matplotlib.pyplot as plt
# Get the unique values in the "Region" column
unique_regions = df["Region"].unique()

# Combine the duplicate values
combined_regions = [f"{region} ({count})" for region, count in df["Region"].value_counts().items()]
# Create the plot
fig = plt.figure(figsize=(15, 5))
plt.bar(unique_regions, combined_regions)
plt.title("Comparison of Regions")
plt.xlabel("REgion")
plt.ylabel("total")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by region and status
grouped_df = df.groupby(["Region", "Status"])

# Count the number of students for each region and status
counts = grouped_df.size()

# Calculate the number of students who are enrolled more than once
counts_more_than_once = counts.unstack(fill_value=0).sum(axis=1) / counts.unstack(fill_value=0).sum(axis=1).max()

# Create the plot
fig, ax = plt.subplots(figsize=(15, 5))
ax.bar(counts.index.levels[0], counts_more_than_once * 100, color=["blue", "red"])
ax.axhline(50, color="black", linestyle="--")

plt.title("Comparison of Regions by Status", fontsize=16)
plt.xlabel("Region", fontsize=14)
plt.ylabel("Persentase", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Group the data by region and status
grouped_df = df.groupby(["Region", "Status"])

# Calculate the average score for each region and status
scores = grouped_df.mean()["INDEPENDENT"]

# Get the level values for the region
regions = scores.index.get_level_values("Region")

# Create the plot
fig, ax = plt.subplots(figsize=(15, 5))
ax.bar(regions, scores)

plt.title("Comparison of Average Student Scores")
plt.xlabel("Region")
plt.ylabel("average INDEPENDENT value ")
plt.tight_layout()
plt.show()
```
